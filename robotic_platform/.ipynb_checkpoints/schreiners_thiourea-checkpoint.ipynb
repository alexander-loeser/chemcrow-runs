```python
import os
from dotenv import load_dotenv
from chemcrow.agents import ChemTools, ChemCrow
from chemcrow.evaluation import NoToolsAgent, Evaluator

load_dotenv()
chemtools = ChemTools()
```

```python
task = """
Find a thiourea organocatalyst which accelerate the Diels-Alder reaction.
After you find it, please plan and execute a synthesis for this organocatalys.
"""

print(task)
```

```python
# First, answer with tools

mrkl = ChemCrow(
    chemtools.all_tools,
    model="gpt-4",
    temp=0.1,
)

result_tools = mrkl.run(task)
```
